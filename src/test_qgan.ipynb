```python
import numpy as np
from scipy.linalg import expm
from numpy.testing import assert_allclose

# Pauli matrices
I = np.eye(2, dtype=complex)
X = np.array([[0, 1], [1, 0]], dtype=complex)
Y = np.array([[0, -1j], [1j, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)
paulis = [I, X, Y, Z]

# Random real coefficients for A and B
np.random.seed(42)
alpha = np.random.uniform(-1, 1, size=4)
beta = np.random.uniform(-1, 1, size=4)

# Build A and B as linear combinations of Pauli matrices
A = sum(a * P for a, P in zip(alpha, paulis))
B = sum(b * P for b, P in zip(beta, paulis))

# Build lifted operators: A  I and I  B
A_lifted = np.kron(A, I)
B_lifted = np.kron(I, B)

# Total Hamiltonian
H = A_lifted + B_lifted

# Compute exponential of the sum
exp_H = expm(H)

# Compute separate exponentials
exp_A = expm(A)
exp_B = expm(B)

# Compute Kronecker product of exponentials
exp_kron = np.kron(exp_A, exp_B)

# Compare
assert_allclose(exp_H, exp_kron, rtol=1e-10, atol=1e-12)
print(" Test passed: exp(AI + IB) == exp(A)exp(B)")
```
